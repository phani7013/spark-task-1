```python
def correlation_heatmap(ss):
    _, ax = plt.subplots(figsize = (20, 10))
    colormap= sns.diverging_palette(220, 10, as_cmap = True)
    sns.heatmap(ss.corr(), annot=True, cmap = colormap)

correlation_heatmap(ss)
```
